```julia
using MixedIntegerExperiments
using Interact
using Plots
using LaTeXStrings
using Polyhedra
using PyPlot
pyplot();
# plotlyjs();
# gr()
```

```julia
v1 = [0.; 0.]
v2 = [1.; 0.]
v3 = [2.; 1.]
vertices = [v1, v2, v3]
P1 = polyhedron(SimpleVRepresentation([v1'; v2']))
P2 = polyhedron(SimpleVRepresentation([v2'; v3']))
polyhedra = [P1, P2];
```

```julia
plt = plot(layout = 2, legend = :none)
plot!(plt[1], xaxis = L"x_1", yaxis = L"x_2")
for (i, v) in enumerate(vertices)
    plot!(plt[1], [v[1]], [v[2]]; markershape = :circle, markercolor = :black, label = latexstring("v^$i")) # TODO labels
end
for (i, P) in enumerate(polyhedra)
    plot_polyhedron!(P, plt[1]; label = latexstring("P^$i"))
    # TODO: labels
end

1s = linspace(0., 1., 10)
2s = linspace(0., 1., 10)
wireframe!(plt[2], 1s, 2s, (1, 2) -> (x = 1 - 1 - 2; ifelse(x < 0, NaN, x)))
# plt.subplots[2].o[:view_init](azim = 100)
# fig[:view_init](elev = 50, azim = 100)
# fig[:view_init](azim = 1.)
plt
```
